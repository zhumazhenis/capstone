### Sobel Filter

```python
import cv2
import numpy as np
from scipy.signal import convolve2d
from scipy.ndimage import imread
import matplotlib.pyplot as plt
```

```python
# Read image
roadGrayImg = cv2.imread('road.jpg', 0)
```

```python
v1 = np.matrix([1, 2, 1])
v2 = np.matrix([-1, 0, 1])

# Sobel filter matrices
Kx = v1.T * v2
Ky = v2.T * v1

# Apply Sobel gilter
Gx = convolve2d(roadGrayImg, Kx, "same", "symm") 
Gy = convolve2d(roadGrayImg, Ky, "same", "symm")
G = np.sqrt(Gx ** 2 + Gy ** 2)
```

```python
plt.imshow(G, cmap='gray')
plt.show()
```
